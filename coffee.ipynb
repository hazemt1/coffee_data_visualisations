```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset =pd.read_csv('/content/sample_data/drinkMenu.csv')
```

```python
dataset=dataset.drop(['Beverage_category','Beverage_prep',' Total Fat (g)', 'Trans Fat (g) ', 'Saturated Fat (g)', ' Sodium (mg)',
       ' Total Carbohydrates (g) ', 'Cholesterol (mg)', ' Dietary Fibre (g)',' Protein (g) ', 'Vitamin A (% DV) ', 'Vitamin C (% DV)',
       ' Calcium (% DV) ', 'Iron (% DV) ', 'Caffeine (mg)'],axis=1)
```

```python
dataset.drop_duplicates(inplace=True)
dataset.dropna(inplace=True)
```

```python
def pltBar(dataset,X,Y,x_label,y_label):
  x=dataset[X]
  y=dataset[Y]

  X_axis = np.arange(len(x))
  plt.figure(figsize=(50,20))
  plt.plot(X_axis , y)

  plt.xticks(X_axis, x)
  # plt.plot(x, y)

  ax = plt.gca()
  plt.draw()

  for tick in ax.get_xticklabels():
      tick.set_rotation(90)

  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.legend()
  plt.show()
  max =0
  highest =''
  for (x,y) in zip(dataset[X],dataset[Y]):
    if y > max:
      max=y
      highest=x

  print('Highest: '+highest+' '+str(max))
```

```python
pltBar(dataset,'Beverage','Calories','Beverage','Calories')
```

```python
pltBar(dataset,'Beverage',' Sugars (g)','Beverage','Sugar')
```
